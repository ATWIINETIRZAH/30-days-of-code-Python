```python
#first we import the library we shall use to analyse data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

#load the dataset
train_df=pd.read_csv('train.csv')
test_df=pd.read_csv("test.csv")

#we shall then explore the dataset
#we shall display the first few rows of the dataset(both train and test)
train_df.head()
test_df.head()




```

```python

#get a summary of the statistics
train_df.describe()

```

```python
#we shall check for missing values

train_df.isnull().sum()

```

```python
#we shall clean the data
#we shall fill the missing values in the age column with the median age

train_df['Age'].fillna(train_df['Age'].median(), inplace=True)
train_df.isnull().sum()
```

```python
#we shall fill the missing values in the Embarked column with the most common port
train_df['Embarked'].fillna(train_df['Age'].median(), inplace=True)
train_df.isnull().sum()
```

```python
#we shall drop the cabin column due to too many missing values
train_df.drop('Cabin',axis=1, inplace=True )
```

```python
#We shall analyze the Data
# first we shall, plot the distribution of survivors
sns.countplot(x='Survived', data=train_df)
plt.title('Distribution of Survivors')
plt.show()
```

```python
#we shall plot the distribution of age
sns.histplot(train_df['Age'], bins=30,kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
#we shall analyze the survival rate by Gender
sns.barplot(x='Sex',y='Survived', data=train_df)
plt.title('Survival Rate by Gender')
plt.show()
```
